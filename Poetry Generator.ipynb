#  NLP Poetry Generation 


```python
import pandas as pd
import numpy as np
import spacy
import codecs
from collections import defaultdict
import random
```

# Data Reading

```python
#Data Readig
path="C:/Users/Umair/Desktop/i21-2081 NLP Assignment 3/"
ghalib="ghalib.txt"
ghalib_data = codecs.open(path+"ghalib.txt", encoding='utf-8').read()
faiz_data = codecs.open(path+"faiz.txt", encoding='utf-8').read()
iqbal_data = codecs.open(path+"iqbal.txt", encoding='utf-8').read()
data=ghalib_data+faiz_data+iqbal_data
print(len(ghalib_data))
print(len(iqbal_data))
print(len(faiz_data))
print(len(data))
```

# Data Cleaning

```python
#Data Cleaning
lst= list()
punctuations = '''‘٪،۔!۔()“ء-[]{};:'"؟\,<>./?@#$’%^&*_~'''
cleaned_data = ""
for x in data:
    x = x.replace('\r\/',' ')
    x = x.replace('\n\/',' ')
    x = x.replace('\n',' ')
    x = x.replace('\r',' ')
    if x not in punctuations:
        cleaned_data = cleaned_data + x

nlp = spacy.blank('ur')
doc = nlp(cleaned_data)
for x in doc:
    lst.append(x)
lst = [str(x).strip(' ') for x in lst]
lst=[x for x in lst if len(x)>0]
```

# UniGram Model

```python
#Frequency Count for unigrams
wordsFreq = defaultdict(int)
for word in lst:
    wordsFreq[word]+=1
uni_gram_freq=pd.DataFrame(wordsFreq.items(), columns=['Uni_Grams', 'Freq'])
uni_gram_freq=uni_gram_freq.sort_values('Freq', ascending=False)
uni_gram_freq.to_csv(r'unigrams.csv')
uni_gram_freq
```

# BiGram Model

```python
#Frequency Count for bigram
bigrams = list()
for i in range(len(lst)-1):
    w1 = lst[i]
    w2 = lst[i+1]
    bigrams.append(tuple((w1, w2)))

wordsFreq = defaultdict(int)
for word in bigrams:
    wordsFreq[str(word[0])+" "+str(word[1])]+=1
bi_gram_freq=pd.DataFrame(wordsFreq.items(), columns=['Bi_Grams', 'Freq'])
bi_gram_freq=bi_gram_freq.sort_values('Freq', ascending=False)
bi_gram_freq.to_csv(r'bigrams.csv')
bi_gram_freq
```

# TriGram Model

```python
#Frequency Count for trigram
trigrams = list()
for i in range(len(lst)-2):
    w1 = lst[i]
    w2 = lst[i+1]
    w3 = lst[i+2]
    trigrams.append(tuple((w1, w2,w3)))

wordsFreq = defaultdict(int)
for word in trigrams:
    wordsFreq[str(word[0])+" "+str(word[1])+" "+str(word[2])]+=1
tri_gram_freq=pd.DataFrame(wordsFreq.items(), columns=['tri_Grams', 'Freq'])
tri_gram_freq=tri_gram_freq.sort_values('Freq', ascending=False)
tri_gram_freq.to_csv(r'trigrams.csv')
tri_gram_freq
```

# Backword BiGram Model

```python
#Frequency Count for bigram
backword_bigrams = list()
for i in range(len(lst)-1):
    w1 = lst[i]
    w2 = lst[i+1]
    backword_bigrams.append(tuple((w1, w2)))

wordsFreq = defaultdict(int)
for word in backword_bigrams:
    wordsFreq[str(word[1])+" "+str(word[0])]+=1
backword_bigrams_freq=pd.DataFrame(wordsFreq.items(), columns=['backword_bigrams', 'Freq'])
backword_bigrams_freq=backword_bigrams_freq.sort_values('Freq', ascending=False)
backword_bigrams_freq.to_csv(r'backword bigrams.csv')
backword_bigrams_freq
```

# Bouns Top Rhyming words and Starting words


```python
lst= list()
punctuations = '''‘٪،۔!۔()“ء-[]{};:'"؟\,<>./?@#$’%^&*_~'''
cleaned_data = ""
for x in data:
    if x not in punctuations:
        cleaned_data = cleaned_data + x
nlp = spacy.blank('ur')
doc = nlp(cleaned_data)
for x in doc:
    lst.append(x)
lst = [str(x).strip(' ') for x in lst]
lst=[x for x in lst if len(x)>0]

top_rhy_words = defaultdict(int)
top_start_words = defaultdict(int)
for index, x in enumerate(lst):
    if index+1 < len(lst) and index-1>=0:
        if str(x) not in punctuations:
            if "\n" in str(x) or "\r" in str(x):
                top_rhy_words[str(lst[index-1])]+=1
                top_start_words[str(lst[index+1])]+=1
    
top_start_words=sorted(top_start_words.items(), key=lambda x: x[1], reverse=True)  
top_rhy_words=sorted(top_rhy_words.items(), key=lambda x: x[1], reverse=True)
```

# BiGram Model Loading and Poetry Generator 


```python
def get_next_bigram_prediction(word):
    bigram_FreqDist = pd.read_csv (r'bigrams.csv')
    for row in bigram_FreqDist["Bi_Grams"]:
        lst=row.split(" ")
        if lst[0]==word:
            return lst[1]

Poem=""
for stanza in range(3):
    rhy_word_index=random.randrange(0, 50)
    last_words=top_rhy_words[rhy_word_index][0]
    for verse in range(4):
        start_word_index=random.randrange(20, 50)
        start_word=top_start_words[start_word_index][0]
        Poem+=str(start_word)+" "
        verse_range=random.randrange(7, 10)
        for x in range(verse_range-2):
            prediction=get_next_bigram_prediction(start_word)
            start_word=prediction
            Poem+=str(prediction)+" "
        Poem+=last_words
        Poem+="\n"
    Poem+="\n\n"
print(Poem)
```

# TriGram Model Loading and Poetry Generator 


```python
def get_next_trigram_prediction(two_word):
    bigram_FreqDist = pd.read_csv (r'trigrams.csv')
    for row in bigram_FreqDist["tri_Grams"]:
        lst=row.split(" ")
        if str(lst[0]) in str(two_word) and str(lst[1]) in str(two_word):
            return lst[2]
Poem=""
for stanza in range(3):
    rhy_word_index=random.randrange(0, 30)
    last_words=top_rhy_words[rhy_word_index][0]
    for verse in range(4):
        first_word_index=random.randrange(20, 50)
        first_word=top_start_words[first_word_index][0]
        sec_word_index=random.randrange(20, 50)
        sec_word=top_start_words[sec_word_index][0]
        Poem+=str(first_word)+" "+str(sec_word)+" "
        verse_range=random.randrange(7, 10)
        for x in range(verse_range-2):
            prediction=get_next_trigram_prediction(str(first_word)+" "+str(sec_word))
            if str(prediction) !="None":
                first_word=sec_word
                sec_word=prediction
                Poem+=str(prediction)+" "
                     
        Poem+=last_words
        Poem+="\n"
    Poem+="\n\n"
print(Poem)
```

```python
def get_next_backword_bigram_prediction(word):
    bigram_FreqDist = pd.read_csv (r'backword bigrams.csv')
    for row in bigram_FreqDist["backword_bigrams"]:
        lst=row.split(" ")
        if lst[0]==word:
            return lst[1]

Poem=""
for stanza in range(3):
    rhy_word_index=random.randrange(0, 50)
    last_words=top_rhy_words[rhy_word_index][0]
    for verse in range(4):
        start_word_index=random.randrange(20, 50)
        start_word=top_start_words[start_word_index][0]
        Poem+=str(start_word)+" "
        verse_range=random.randrange(7, 10)
        for x in range(verse_range-2):
            prediction=get_next_backword_bigram_prediction(start_word)
            start_word=prediction
            Poem+=str(prediction)+" "
        Poem+=last_words
        Poem+="\n"
    Poem+="\n\n"
print(Poem)
```
